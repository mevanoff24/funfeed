```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

train = pd.read_csv('data/new_train.csv')
test = pd.read_csv('data/new_test.csv')
train.head()
```

```python
def create_baselines_test(feature, name):

    gb = train.groupby(feature).mean()
    dat = pd.merge(gb, test[['item', 'user_id']], left_index=feature, right_on=feature, how = 'right')
    dat.rating.fillna(value=dat.rating.mean(), inplace=True)
    dat = dat[['item', 'user_id', 'rating']]
#     dat.to_csv('data/' + name + '_mean_benchmark.csv', header = None, index = False)
    return dat
```

```python
item_benchmark = create_baselines_test(['item'], 'item')
user_benchmark = create_baselines_test(['user_id'], 'user')
user_and_item_benchmark = create_baselines_test(['item', 'user_id'], 'item_and_user')
```

```python
user_benchmark
```

```python
dat_join = item_benchmark.join(user_benchmark, lsuffix="l_", rsuffix='r_')
dat_join.head()
```

```python
y_pred_item = dat_join.ratingl_.values
y_pred_user = dat_join.ratingr_.values
```

```python
a = np.column_stack((y_pred_item, y_pred_user))
pred = a.mean(axis = 1)
```
